## BROADCASTING IN AN ARRAY
1. Broadcasting lets Numpy perform element wise operations on array of different shapes.
2. It automatically expands smaller arrays to match with larger one's.
3. This makes computation faster and more efficient without writing loops.

```python
a = [10,20,30,40,50]
b = 100
```

```python
result = []
```

```python
for x in a:
    result.append(x+b)
print(result)
```

```python
a_1 = np.array([10,20,30,40,50])
b_1 = 100
```

```python
print(a_1+b_1)
```

## STATISTICAL METHODS
1. MEAN
2. MEDIAN
3. MODE

```python
# Generate random integers in the range of 100-200 and also should be in 5*5 matrix
rand_8 = np.random.randint(100,200,(5,5))
rand_8
```

```python
print(np.mean(rand_8))
print(np.median(rand_8))
print(np.std(rand_8))
```

```python
## MIN/MAX
arr_9 = np.random.randint(100,500,20)
arr_9
```

```python
print(np.min(arr_9)) # returns the smallest number
print(np.max(arr_9)) # returns the largest number
```

```python
print(np.argmin(arr_9))  # returns the index location of smallest number
print(np.argmax(arr_9))  # returns the index location of largest number
```

## FILTERING
-- Returns the values based on a condition.

```python
arr2 = np.array([1,2,1010,4,108,18,71,610])
arr2
```

```python
arr2<100
```

```python
np.where(arr2<100) # returns the index location where the condition is true
```

```python
arr2[np.where(arr2<100)] # returns the actual values that satisfies the condition
```

```python
arr3 = np.array([100,105,200,205,150,140,300,500,5000,1000,670,1000])
arr3
```

```python
arr3[np.where(arr3>500)]
```

```python
np.where(arr3>500)
```

```python
arr3[np.where(arr3==400)]=500  # changing values based on condition
```

```python
arr3
```

```python
# find the index location of all 400
np.where(arr3==400)
```

## Fancy Indexing
-- Fancy indexing means accessing array elements using list of arrays of indices, instead
 of a single index or slice.

-- It helps in retrieving elements in any order and is useful do advanced selection.

```python
fan_arr = np.arange(100).reshape(10,10)
fan_arr
```

```python
fan_arr[[6,3,9,1]]  # fancy indexing on rows
```

```python
# fancy indexing on columns 2-92,5-95,7-97
fan_arr[:,[2,5,7]]
```
